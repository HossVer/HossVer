## <strong>Memulai</strong>
</div>

Dokumen yang sedang Anda baca bukan halaman statis. Ini media interaktif bernama <strong>notebook Colab</strong> yang memungkinkan Anda menulis dan mengeksekusi kode.

Misalnya, berikut ini <strong>kode sel</strong> dengan skrip Python pendek yang mengomputasi sebuah nilai, menyimpannya dalam suatu variabel, dan mencetak hasilnya:

```
#sjgg output jamak
import matplotlib.pyplot as plt
import numpy as np

def TrainingPersceptron(x, T):
  #1. inisiasi Data
  JumlahData = x.shape[0]
  JumlahVektorInput = x.shape[1]
  JumlahVektorOutput = T.shape[1]

  #2.inisasi bobot Data
  w=np.random.rand(JumlahVektorInput, JumlahVektorOutput)*0.5

  #3.Menentukan Learning Rate sebesar 0.5
  alpha = 0.5
  loss = []

  #4.Melakukan iterasi untuk tuning bobot sebanyak 1000 kali
  for epoh in range(1000):
    loss.append(0)

  #5 Tunning bobot untuk setiap data
    for NData in range(JumlahData):
      xi = x[NData: NData+1, :]
      net = np.matmul(xi,w)
      #fungi aktivasi sigmoid untuk mencari output
      y = 1/ (1+ np.exp(-net))
      #menghitung Eror Propagasi
      e = T[NData: NData+1, :]-y
      #menghitung perubahan bobot
      dw = alpha*np.matmul(xi.transpose(),e)
      w = w+dw
      #menghitung loss dengan fungsi sum error
      loss[epoh]=loss[epoh] + np.sum(np.abs(e))
    return w,loss

def Prediksi(xp,w):
  net = np.matmul(xp,w)
  #menerapkan fungsi aktivasi sigmoid
  yp = 1/ (1+ np.exp(-net))
  return yp;

#1.Main Program
#1. Pembelajaran JST Perceptron
#Target xJamak.csv
#Input tJamak.csv

#mebaca file inputan X jari xJamak.csv
X = np.loadtxt("XJamak.csv", delimiter=',')
#mebaca file target X jari tJamak.csv
T = np.loadtxt("TJamak.csv", delimiter=',')

#Melakukan pembelajaran Perceptron
w,loss = TrainingPersceptron(X,T)

#Visualisasi grafik loss
plt.plot(loss)

#2.Prediksi berdasar bobot hasil training
xp = np.array([[1,0,1,1]])
yp = Prediksi(xp,w)

print("Hasil Prediksi ")
print(yp)
```

```
import numpy as np #Mengolah data, merubah list menjadi array
from keras.models import Sequential #Ambil fungsi sequentian untuk fungsi add()
                                    #untuk penambahan layer ke mod
from keras.layers import Dense #Membuat layer dense pada model
import matplotlib.pyplot as plt #Visualiasi sebaran data & tabel

#1. Siapkan dataset X dan Y, X sebagai input Y sebagai Target/Output yang ingin dicapai
X = np.array([[0,0],
             [0,1],
             [1,0],
             [1,1]],
             )

T = np.array([[1],
             [0],
             [0],
             [0]],
             )

#Pada JST terdapat bias yang digunakan untuk membantu menentukan aktivasi sigmoid. 
#2. Definisikan model Jst
model = Sequential()
#. Menambahkan Layer Dense ke Model
NNode = 1 #Jumlah Node
NInput = 2 #Jumlah Vektor Input
model.add(Dense(NNode, input_dim=NInput, activation='sigmoid',use_bias="True"))

#3. Compile model
model.compile(loss='mse', optimizer='SGD',metrics=['accuracy'])

#4. Training Model dengan 4000 epoh (Jumlah Iteasi)
#Input = X ||  Target = Y
His = model.fit(X,T, epochs=4000)

#Visualiasi grafik loss
plt.plot(His.history['loss'])

#5. Terapkan untuk prediksi (X)
yp = model.predict(X)
print(yp)
```
